```python
import pandas as pd
```

```python
df = pd.read_csv("../data/input/Cleaned_DS_jobs.csv")
```

```python
jd = df['Job Description'][1]
```

```python
keywords = list(df.columns[-9:-2])
```

```python
keywords.append('sql')
keywords.append('mysql')
keywords.append('my sql')
keywords.append('power bi')
```

```python
keywords[-5] = "big data"
```

```python
jds = df['Job Description']
```

```python
kws = []
for job in jds:
    kw = []
    for i in keywords:
        if i in job.lower():
            kw.append(i)
    kws.append(kw)
kws
    
```
